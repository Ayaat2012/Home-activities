```python
import seaborn as sb
import matplotlib.pyplot as plt

# Checking the distribution of some features:-
df = pd.read_csv('Tips Dataset.csv')

# To visualize the distribution of 'total_bill':-
sb.displot(df['total_bill'], kde = True)
plt.show()

# To visualize the distribution of 'size':-
sb.displot(df['size'], kde=True)
plt.show()
# To visualize the distribution of 'tip':-
sb.displot(df['tip'], kde=True)
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Creating a histogram with KDE parameters:-
sb.histplot(df['total_bill'], kde=True)
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Creating a scatter plot:-

plt.figure(figsize=(8, 6))

sb.scatterplot(data = df, x = "total_bill", y = "tip", hue = "time")
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Creating a pair plot:-
sb.pairplot(df, hue = "gender")
plt.show()
```
