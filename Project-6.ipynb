```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('IMDB Dataset.csv')

# Creating a boxplot of IMDB_Rating and Runtime
sns.boxplot(x='IMDB_Rating', y='Runtime', data=df)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('IMDB Dataset.csv')

# Checking the relationship between IMBD_Rating and Runtime
plt.scatter(x=df['IMDB_Rating'], y=df['Runtime'])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('IMDB Dataset.csv')

# Checking the distribution of IMBD_Rating and Runtime
sns.distplot(df['IMDB_Rating'])
plt.show()
sns.distplot(df['Runtime'])
plt.show()
```
